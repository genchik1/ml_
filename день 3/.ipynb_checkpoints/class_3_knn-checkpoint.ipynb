# Выбор числа соседей в метода kNN

```python
# хранине данных
import pandas as pd
import numpy as np

# Обучене модели и подготовка данных
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.preprocessing import scale
from sklearn.model_selection import cross_val_score

# визуализация
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
sns.set(rc={'figure.figsize':(12,8)})

import warnings
warnings.filterwarnings('ignore')
```

## Загружаем данные

```python
df = pd.read_csv('./wine.data', header=None, names=['TARGET', 'Alcohol', 'Malic_acid',
'Ash',
'Alcalinity_of_ash'  
,'Magnesium'
,'Total_phenols'
,'Flavanoids'
,'Nonflavanoid_phenols'
,'Proanthocyanins'
,'Color_intensity'
,'Hue'
,'OD_wines'
,'Proline'])
```

```python
df.TARGET.value_counts()
```

```python
sns.distplot(df.TARGET)
```

```python
df.head()
```

#### выделяем целевую переменную и набор признаков

```python
Y = df['TARGET']
X = df[list(set(df.columns).difference(['TARGET']))]
```

#### Создаем генератор разбиения и рпоизводим обучение метода kNN

```python
kf = KFold(n_splits=5, random_state=42, shuffle=True)
knn = KNeighborsClassifier(n_neighbors=30)
scores = cross_val_score(knn, X, Y, cv=10, scoring='accuracy')

print("Среднее качество: {0:.2f}%".format(scores.mean()*100))
print("Стандартное отклонение: {0:.2f}%".format(scores.std()*100))
```

#### Подбор параметра k-числа ближайших соседей методом cross-validation

```python
cv_scores=[]
for k in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=5, scoring='accuracy')
    cv_scores.append([k, scores.mean(),scores.std()])
df_scores = pd.DataFrame(cv_scores, columns=['k', 'ACCURACY', 'STD'])
```

```python
sns.lineplot(x="k", y="ACCURACY", data=df_scores)
```

```python
X_scaled = scale(X)
```

```python
cv_scores_sc=[]
for k in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_scaled, Y, cv=5, scoring='accuracy')
    cv_scores_sc.append([k, scores.mean(),scores.std()])
df_scores_sc = pd.DataFrame(cv_scores_sc, columns=['k', 'ACCURACY', 'STD'])
```

```python
sns.lineplot(x="k", y="ACCURACY", data=df_scores_sc)
```

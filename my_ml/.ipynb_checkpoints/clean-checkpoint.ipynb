```python
df = data.copy()
```

```python
print (len(df))
# df["WORK_TIME"] = df["WORK_TIME"].map({np.nan:0})
df = df.dropna(subset=["WORK_TIME"])
print (len(df))
```

```python
def clean_work_time(x):
    if (x['AGE']-12)*12 >= x['WORK_TIME']:
        return True
    else:
        return False
    
df['y'] = df.apply(lambda x: clean_work_time(x), axis=1)
df = df[df['y']==True].drop('y', axis=1)
print (len(df))
```

```python
df = df.dropna()
print (len(df))
print (data['TARGET'].value_counts(normalize=True))
print (df['TARGET'].value_counts(normalize=True))
```

```python
object_column = [x for x in df.columns if df[x].dtypes == object]
print (object_column)
df[object_column].head()
```

```python
def _rename(df, x, uniq):
    df[x] = df[x].map(uniq)
    return df
def rename(df, x):
    uniq = df[x].unique()
    uniq = dict(enumerate(uniq))
    uniq = dict(map(reversed, uniq.items()))
    df = _rename(df, x, uniq)
    return df, uniq
```

```python
for x in ['GEN_INDUSTRY', 'GEN_TITLE', 'ORG_TP_STATE', 'ORG_TP_FCAPITAL', 'JOB_DIR']:
    df, _ = rename(df, x)
```

```python
df, uniq = rename(df, 'REG_ADDRESS_PROVINCE')
```

```python
for x in ['FACT_ADDRESS_PROVINCE', 'POSTAL_ADDRESS_PROVINCE']:
    df = _rename(df, x, uniq)
```

```python
object_column = [x for x in df.columns if df[x].dtypes == object]
print (object_column)
df[object_column].head()
```

```python
for x in ['TP_PROVINCE', 'REGION_NM']:
    df, _ = rename(df, x)
```

```python
for x in ['CREDIT', 'FST_PAYMENT', 'LOAN_AVG_DLQ_AMT', 'LOAN_MAX_DLQ_AMT', 'PERSONAL_INCOME']:
    df[x] = df[x].str.replace(',', '.').astype(float)
```

```python
df['FAMILY_INCOME'] = df['FAMILY_INCOME'].str.split(' руб').str[0]
df['FAMILY_INCOME'] = df['FAMILY_INCOME'].str.split(' ').str[-1]
```
